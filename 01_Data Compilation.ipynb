### Run the following on loop
<br>Create date time for num data set using cat data set
<br>Take value counts of date time in both data sets
<br>Select only that date times for whcih counts are matching for both the data sets
<br>Merge the data sets and run round of checks to see if the merge is successful

```python
for i in [117]:
    df=pd.DataFrame()
    df1=pd.read_csv(r"data\csh"+str(i)+"\csh"+str(i)+".ann.features.csv")
    df2=pd.read_csv(r"data\csh"+str(i)+"\csh"+str(i)+".ann.txt",sep='\t',header=None,names=['date_time','sensor','location','location2','message','senor_type','activity2'])
    print('csh'+str(i)+' df1 size: '+ str(df1.shape[0]))
    print('csh'+str(i)+' df2 size: '+ str(df2.shape[0]))
    
    part_amt=round(df1.shape[0]/20)
    df1_dict={}
    df2_dict={}
    for j in range(11,20):
        if j!=19:
            df1_dict[j]=df1[j*part_amt:j*part_amt+part_amt]
            df2_dict[j]=df2[j*part_amt:j*part_amt+part_amt]
        else:
            df1_dict[j]=df1[j*part_amt:]
            df2_dict[j]=df2[j*part_amt:]
    for p in range(11,20):
        df1=df1_dict[p].reset_index(drop=True)
        df2=df2_dict[p].reset_index(drop=True)
        print('csh'+str(i)+'_'+str(p+1)+' df1 size: '+ str(df1.shape[0]))
        print('csh'+str(i)+'_'+str(p+1)+' df2 size: '+ str(df2.shape[0]))

        df1['time']=df1.apply(lambda x: "{0:0=2d}".format(round(x['lastSensorEventHours']))+":{0:0=2d}".format(floor((x['lastSensorEventSeconds']/60-x['lastSensorEventHours']*60)))+":{0:0=2d}".format(int(x['lastSensorEventSeconds']-x['lastSensorEventHours']*60*60-floor((x['lastSensorEventSeconds']/60-x['lastSensorEventHours']*60))*60)),axis=1)
        df2['date_time']=df2['date_time'].apply(lambda x: x[:-7])
        df2['time2']=df2['date_time'].apply(lambda x: x[11:])
        df2.loc[df2['time2'].apply(lambda x: len(x))!=8,'time2']='00:00:00'
        df2=df2.drop_duplicates()

        j=2
        min_df1_val=pd.to_datetime(df1['time'],format='%H:%M:%S').drop_duplicates().iloc[:50].nsmallest(j).iloc[-1]
        min_df2_val=pd.to_datetime(df2['time2'],format='%H:%M:%S').drop_duplicates().iloc[:50].nsmallest(j).iloc[-1]
        threshold=str(max([min_df1_val,min_df2_val]))[11:]

        df1=df1.iloc[df1['time'].iloc[:100][df1['time'].iloc[:100]>=threshold].index[0]:].reset_index(drop=True)
        df2=df2.iloc[df2['time2'].iloc[:100][df2['time2'].iloc[:100]>=threshold].index[0]:].reset_index(drop=True)
        print('csh'+str(i)+'_'+str(p+1)+' df1 size: '+ str(df1.shape[0]))
        print('csh'+str(i)+'_'+str(p+1)+' df2 size: '+ str(df2.shape[0]))

        if df1.shape[0]!=df2.shape[0]:
            temp_df=pd.concat([pd.DataFrame(df1[df1['lastSensorEventHours'].diff(1)<0].index),(pd.Series({0:df1.shape[0]}))]).reset_index(drop=True)
            temp_df['start']=pd.concat([pd.Series({0:0}),(temp_df[0].iloc[:-1])]).reset_index(drop=True)
            date_list = [datetime.datetime.strptime(df2['date_time'].iloc[0][:10], "%Y-%m-%d")]
            for l in range(temp_df.shape[0]-1):
                date_list.append(date_list[-1]+ datetime.timedelta(days=1))
            temp_df['date']=date_list
            df1['date']=0
            for k in range(temp_df.shape[0]):
                df1.loc[temp_df['start'].iloc[k]:temp_df[0].iloc[k],'date']=temp_df.loc[k,'date']
            df1['date_time1']=df1.apply(lambda x: str(x['date'])[:11]+x['time'],axis=1)
            temp_df2=pd.merge(df2['date_time'].value_counts().reset_index(),df1['date_time1'].value_counts().reset_index(),on='index',how='outer')
            exclude_list=list(temp_df2[temp_df2['date_time']!=temp_df2['date_time1']]['index'])
            df1=df1[~(df1['date_time1'].apply(lambda x: x in exclude_list))].reset_index(drop=True)
            df2=df2[~(df2['date_time'].apply(lambda x: x in exclude_list))].reset_index(drop=True)
            print('csh'+str(i)+'_'+str(p+1)+' df1 size: '+ str(df1.shape[0]))
            print('csh'+str(i)+'_'+str(p+1)+' df2 size: '+ str(df2.shape[0]))

        if df1.shape[0]==df2.shape[0]:
            print('We can successfully merge the data for dataset: csh'+str(i)+'_'+str(p+1))
            df=pd.concat([df2,df1],axis=1)
        else:
            print('Please match the rows to merge the data for dataset: csh'+str(i)+'_'+str(p+1))
            pass
        if df.shape[0]>0:
            if (df[df['activity2']!=df['activity']].shape[0]==0) and (df[df['time2']!=df['time']].shape[0]==0):
                print('Data Merge Successfull for dataset: csh'+str(i)+'_'+str(p+1))
                df['room']="csh"+str(i)
                df.to_csv(r"data\csh"+str(i)+'_'+str(p+1)+"_combined.csv",index=None)
            else:
                print('Please check the merge for dataset: csh'+str(i)+'_'+str(p+1))
                temp_df2.to_csv(r"data\csh"+str(i)+'_'+str(p+1)+"_temp_df2.csv",index=None)
                temp_df.to_csv(r"data\csh"+str(i)+'_'+str(p+1)+"_temp_df.csv",index=None)
                pass
        else:
            print('Please check the df for dataset: csh'+str(i)+'_'+str(p+1))
            pass
```
